```python
import torch
import matplotlib.pyplot as plt

# Example tensor of size (4, 4, 4)
tensor = torch.load('tensor_list.pt')
print(len(tensor))
# Plot slices along one axis (for example, axis 0)
point_cloud = tensor[0].cpu().data.numpy()
```

```python
import plotly.graph_objects as go
import numpy as np
x = point_cloud[:, :, 0].ravel()
y = point_cloud[:, :, 1].ravel()
z = point_cloud[:, :, 2].ravel()

# print(point_cloud)
# Create a 3D scatter plot
fig = go.Figure(data=[go.Scatter3d(x=x, y=y, z=z, mode='markers')])

# Update layout with axis labels and title
fig.update_layout(scene=dict(
    xaxis_title='X',
    yaxis_title='Y',
    zaxis_title='Z'),
    title="3D Point Cloud Visualization"
)
```

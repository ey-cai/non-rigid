```python
import rpad.visualize_3d.plots as vpl
import os
import zarr
import torch
import numpy as np

train_zarr = zarr.open('/home/ktsim/Projects/non-rigid/data/proccloth/cloth=single-fixed anchor=single-random hole=single/train.zarr')
        # goalPC_path = os.path.join(self.zarr_dir, 'tensor_list_train.pt')

tensor = torch.load('tensor_list_train.pt')
# Plot slices along one axis (for example, axis 0)
goal_point_cloud = tensor[0].cpu().data.numpy().squeeze(0)

fig = vpl.segmentation_fig(
    np.concatenate([
        train_zarr['data/action_pcd'][0],
        train_zarr['data/anchor_pcd'][0],
        goal_point_cloud,
    ]),
    np.concatenate([
        np.ones(train_zarr['data/action_pcd'][0].shape[0]),
        np.zeros(train_zarr['data/anchor_pcd'][0].shape[0]),
        np.ones(goal_point_cloud.shape[0]) * 2,
        
    ]).astype(int)
)
fig.show()
```

```python
import rpad.visualize_3d.plots as vpl
import os
import zarr
import torch
import numpy as np

train_zarr = zarr.open('/home/ktsim/Projects/non-rigid/data/old_dataset/cloth=single-fixed anchor=single-random hole=single/train.zarr')
episode_num = 0
# print(train_zarr['data/ground_truth'][0].shape)
print(train_zarr['data/force'].shape)
fig = vpl.segmentation_fig(
    np.concatenate([
        train_zarr['data/action_pcd'][episode_num],
        train_zarr['data/anchor_pcd'][episode_num],
        train_zarr['data/ground_truth'][episode_num],
        train_zarr['data/force'][:, 0, :] + np.array([3, 5, 10]),
    ]),
    np.concatenate([
        np.ones(train_zarr['data/action_pcd'][episode_num].shape[0]),
        np.zeros(train_zarr['data/anchor_pcd'][episode_num].shape[0]),
        np.ones(train_zarr['data/tax3d'][episode_num].shape[0],) * 2,
        np.array([1]*60 + [2]* 60 + [3]* 60 + [6]*121),
        
    ]).astype(int)
)
fig.show()

```
